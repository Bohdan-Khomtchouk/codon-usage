```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vrl_df = pd.read_csv("../heatmap_inputs/vrl_data.csv", index_col=0)
phg_df = pd.read_csv("../heatmap_inputs/phg_data.csv", index_col=0)
bct_df = pd.read_csv("../heatmap_inputs/bct_data.csv", index_col=0)
arc_df = pd.read_csv("../heatmap_inputs/arc_data.csv", index_col=0)
euk_df = pd.read_csv("../heatmap_inputs/euk_data.csv", index_col=0)
```

```python
vrl_df.head(5)
```

```python
#Heatmaps without normalizing 

fig,(ax1,ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(10,10))

g1 = sns.heatmap(vrl_df, cbar=False, ax=ax1)
ax1.title.set_text('vrl')
g2 = sns.heatmap(phg_df, cbar=False, ax=ax2)
ax2.title.set_text('phg')
g3 = sns.heatmap(bct_df, cbar=False, ax=ax3)
ax3.title.set_text('bct')
g4 = sns.heatmap(arc_df, cbar=False, ax=ax4)
ax4.title.set_text('arc')
g5 = sns.heatmap(euk_df, cbar=False, ax=ax5)
ax5.title.set_text('euk')

fig.tight_layout()
```

```python
#Heatmaps after normalizing 
fig,(ax1,ax2, ax3, ax4, ax5) = plt.subplots(5, 1, figsize=(10,10))

g1 = sns.heatmap(vrl_df,vmin=0, vmax=0.27543, cbar=False, ax=ax1)
ax1.title.set_text('vrl')
g2 = sns.heatmap(phg_df,vmin=0, vmax=0.27543, cbar=False, ax=ax2)
ax2.title.set_text('phg')
g3 = sns.heatmap(bct_df,vmin=0, vmax=0.27543, cbar=False, ax=ax3)
ax3.title.set_text('bct')
g4 = sns.heatmap(arc_df,vmin=0, vmax=0.27543, cbar=False, ax=ax4)
ax4.title.set_text('arc')
g5 = sns.heatmap(euk_df,vmin=0, vmax=0.27543, cbar=False, ax=ax5)
ax5.title.set_text('euk')

fig.tight_layout()
```
